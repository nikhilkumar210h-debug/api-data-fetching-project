```python

frames = []   # list to store each page’s DataFram

for i in range(1, 501):
    url = f"https://api.themoviedb.org/3/movie/top_rated"
    params = {
        "api_key": "183099375f1634a4a8aa383f04afae2c",
        "language": "en-US",
        "page": i
    }
    res = requests.get(url,params=params, timeout=10)
    data = res.json()

    if "results" in data:
        temp_df = pd.DataFrame(data["results"])[[
            "id", "adult", "original_title", "release_date",
            "vote_count", "popularity", "overview"
        ]]
        frames.append(temp_df)  
    else:
        print(f"⚠️ No results on page {i}: {data}")


df = pd.concat(frames, ignore_index=True)

print("Total movies collected:", len(df))
print(df.head())
```
